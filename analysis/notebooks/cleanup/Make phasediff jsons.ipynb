```python
from bids import BIDSLayout
```

```python
layout = BIDSLayout('/data/risk_precision/ds-risk/sourcedata/', validate=False)
```

```python
df = layout.to_df()
```

```python
df = df[~df.subject.isnull()]
```

```python
df['acquisition'] = df.acquisition.astype(float)
```

```python
import json
```

```python
for subject, d in df.groupby('subject'):
    
    if (d.suffix == 'bold').any():

        phasediff = d[d.suffix == 'phasediff']

        mapping = {key:[] for key in phasediff.run}

        for run, d2 in d[d.suffix == 'bold'].set_index('run').iterrows():
            ix = (phasediff['acquisition'] - d2.acquisition).abs().argmin()
            phasediff_match = phasediff.loc[ix].run

            mapping[phasediff_match].append(d2.path.replace('/data/risk_precision/ds-risk/sourcedata/sub-{}/'.format(d2.subject),
                                                            ''))
            
        
        
        for ix, row in phasediff.iterrows():
            json_dict = {"EchoTime1": 0.0046, "EchoTime2": 0.0069}
            json_dict['IntendedFor'] = mapping[row.run]
            
            json_fn = row.path.replace('.nii.gz', '.json')
            
            
            print(json_fn)
            
            with open(json_fn, 'w') as f:
                json.dump(json_dict, f)
```
